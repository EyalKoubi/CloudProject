```python

from IPython.utils.text import textwrap


import ipywidgets as widgets
i = 0
arr = ['0','0','0','0']
with open('students.txt', 'r') as s:
  for line in s:
    arr[i] = line
    i+=1

from typing import Text

w = widgets.Dropdown (
    options = arr,
    value = arr[0],
    description = "students",
)

txt1 = widgets.Text()

details = 0

global arrW
arrW = ['0','0','0','0','0','0']

global string
string = ""

global flag
flag = 1

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
      global flag
      global arrW
      global string
      if flag == 0:
        if arrW[4].value != "":
          with open('students.txt', 'a') as file:
            file.write(arrW[0].value+" "+arrW[1].value+" "+arrW[2].value+" "+arrW[3].value+" "+arrW[4].value+" "+arrW[5].value+'\n')
        for i in range(6):
          arrW[i].close()
      flag = 0
      txt1.value = w.value
      details = txt1.value.split()
      arrW[0] = widgets.Text (
          value = details[0],
          description = 'firstName',
          disabled = True,
      )
      arrW[1] = widgets.Text (
          value = details[1],
          description = 'lastName',
          disabled = True,
      )
      arrW[2] = widgets.Text (
          value = details[2],
          description = 'mail',
          disabled = True,
      )
      arrW[3] = widgets.Text (
          value = details[3],
          description = 'course',
          disabled = True,
      )
      arrW[4] = widgets.Text (
          value = details[4],
          description = 'link',
          disabled = True,
      )
      arrW[5] = widgets.Text (
          value = "",
          description = 'new',
          disabled = False,
      )
      for i in range(6):
        display(arrW[i])

w.observe(on_change)

display(w)
```
