```python
# imports
from IPython.utils.text import textwrap
import json
import ipywidgets as widgets
from typing import Text

# define array for the jsons
arr = []

# read the jsons from text file
# and fill the array
with open('students1.txt') as f:
    for jsonObj in f:
        studentDict = json.loads(jsonObj)
        arr.append(studentDict)

# initialize array for our names
new_arr = ['0','0','0','0','0','0']

# fill the array of our names
for i in range(len(arr)):
  new_arr[i] = str(arr[i]).split(",")[0].split(":")[1].replace("'","")

# define the comboBox
w = widgets.Dropdown (
    options = new_arr,
    value = new_arr[0],
    description = "students",
)

# initialize array for details
# of specific person
details = ['0','0','0','0','0','0']

# initialize (global) array for
# details text of specific person
global arrW
arrW = ['0','0','0','0','0','0']

# initialize flag for
# the first comboBox
# appereance
global flag
flag = 1

# define function for each action
# on the comboBox
def on_change(change):
    # define action for specific choice
    if change['type'] == 'change' and change['name'] == 'value':
      global flag
      global arrW

      # if this is not the first
      # comboBox apereance
      if flag == 0:
        # if we didn't choose favorite thing
        if arrW[5].value != "":
          # write to other document the additional things
          with open('students2.txt', 'a') as file:
            file.write('\n\n'+arrW[0].value+" "+arrW[1].value+" "+arrW[2].value+" "+arrW[3].value+" "+arrW[4].value+" "+arrW[5].value+'\n')
        # close the windows of the
        # last comboBox choice
        for i in range(len(arrW)):
          arrW[i].close()

      # update the first apereance flag
      flag = 0

      # get the original json of
      # the specific choice
      for i in range(len(arr)):
        if (w.value == str(arr[i]).split(",")[0].split(":")[1].replace("'","")):
          temp = str(arr[i]).split(',')
          break

      # get the cleaned details
      # from the original jsons
      for i in range(len(temp)):
        details[i] = createCleanString(temp[i])

      # get the details for
      # the texts on the screen

      # first name
      arrW[0] = widgets.Text (
          value = details[0],
          description = 'firstName',
          disabled = True,
      )

      # last name
      arrW[1] = widgets.Text (
          value = details[1],
          description = 'lastName',
          disabled = True,
      )

      # mail
      arrW[2] = widgets.Text (
          value = details[2],
          description = 'mail',
          disabled = True,
      )

      #course
      arrW[3] = widgets.Text (
          value = details[3],
          description = 'course',
          disabled = True,
      )

      # link
      arrW[4] = widgets.Text (
          value = details[4],
          description = 'link',
          disabled = True,
      )

      # additional thing
      arrW[5] = widgets.Text (
          value = "",
          description = 'new',
          disabled = False,
      )

      # display the comboBox
      for i in range(len(arrW)):
        display(arrW[i])

# activate the
# comboBox function
w.observe(on_change)

# display all
display(w)

# define method that clean the
# jsons elements and return
# the details areselfs as array
def createCleanString(str):
  res = ""
  flagDot = False
  for i in range(len(str)):
    if str[i]==":":
      flagDot = True
    if flagDot:
      while str[i] != "'":
        i+=1
      i+=1
      while str[i] != "'":
        res += str[i]
        i+=1
      break
  return res
```
