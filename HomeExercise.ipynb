```python
# imports
from IPython.utils.text import textwrap
import json
import ipywidgets as widgets
from typing import Text

# initialize array for the jsons
arr = []

# define function that read
# the jsons from text file
# and fill the array
def load():
  with open('students1.txt') as f:
      for jsonObj in f:
          studentDict = json.loads(jsonObj)
          arr.append(studentDict)

# define function that creat json line
def writeLine(str1,str2,str3,str4,str5,str6):
  result = ''
  result = '{"first name": "'
  result += str1.replace(" ","")
  result += '", "last name": "'
  result += str2.replace(" ","")
  result += '", "mail": "'
  result += str3.replace(" ","")
  result += '", "proficient": "'
  result += str4.replace(" ","")
  result += '", "link": "'
  result += str5.replace(" ","")
  result += '", "favorite show": "'
  result += str6
  result += '"}'
  return result

# activate load
load()

# initialize array for our names
new_arr = ['0','0','0','0','0','0']

# fill the array of our names
for i in range(len(arr)):
  new_arr[i] = str(arr[i]).split(",")[0].split(":")[1].replace("'","")

# define the comboBox
w = widgets.Dropdown (
    options = new_arr,
    value = new_arr[0],
    description = "students",
)

# initialize array for details
# of specific person
details = ['0','0','0','0','0','0']

# initialize (global) array for
# details text of specific person
arrW = ['0','0','0','0','0','0']

# initialize flag for
# the first comboBox
# appereance
flag = 1

# define function for each action
# on the comboBox
def on_change(change):

    # global variables
    global index
    global buffer
    global arrW
    global flag
    global arr

    # define action for specific choice
    if change['type'] == 'change' and change['name'] == 'value':

      # if this is not the first
      # comboBox apereance
      if flag == 0:

        # if we didn't choose favorite show
        if arrW[5].value != "":
          buffer = ""
          for i in range(len(arr)):
            if (i==index):
                arr[index] = writeLine(arrW[0].value,arrW[1].value,arrW[2].value,arrW[3].value,arrW[4].value,arrW[5].value)
                buffer += '\n' + str(arr[index])
            else:
              buffer += '\n' + str(arr[i])
          # write the new jsons to the text ('students1.txt')
          with open('students1.txt', 'w') as file:
            file.write(buffer.replace('"',"'"))
        # close the windows of the
        # last comboBox choice
        for i in range(len(arrW)):
          arrW[i].close()

      # update the first apereance flag
      flag = 0

      # get the original json of
      # the specific choice
      for i in range(len(arr)):
        if (w.value.replace(" ","") == str(arr[i]).split(",")[0].split(":")[1].replace("'","").replace('"','').replace(" ","")):
          index = i
          temp = str(arr[i]).split(',')
          break

      # get the cleaned details
      # from the original jsons
      for i in range(len(temp)):
        details[i] = temp[i].split(':')[1].replace('"','').replace('{','').replace('}','').replace("'","")

      # get the details for
      # the texts on the screen

      # first name
      arrW[0] = widgets.Text (
          value = details[0],
          description = 'firstName',
          disabled = True,
      )

      # last name
      arrW[1] = widgets.Text (
          value = details[1],
          description = 'lastName',
          disabled = True,
      )

      # mail
      arrW[2] = widgets.Text (
          value = details[2],
          description = 'mail',
          disabled = True,
      )

      # course
      arrW[3] = widgets.Text (
          value = details[3],
          description = 'course',
          disabled = True,
      )

      # link
      arrW[4] = widgets.Text (
          value = details[4],
          description = 'link',
          disabled = True,
      )

      # favorite tv show
      arrW[5] = widgets.Text (
          value = "",
          description = 'tvShow',
          disabled = False,
      )

      # display the comboBox
      for i in range(len(arrW)):
        display(arrW[i])

# activate the
# comboBox function
w.observe(on_change)

# display all
display(w)
```
